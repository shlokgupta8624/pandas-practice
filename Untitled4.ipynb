```python
patients_df = patients.copy()

patients_df['email'] = patients_df['contact'].str.extract(
    r'([\w\.-]+@[\w\.-]+\.\w+)'
)

patients_df['phone'] = patients_df['contact'].str.extract(
    r'(\+?\d[\d\s\-\(\)]{7,}\d|\d{3}-\d{3}-\d{4})'
)

patients_df[['contact', 'email', 'phone']].head()
```
